```python
import pandas as pd
oversampling_data = pd.read_csv("oversampling result.csv",index_col = 0)
oversampling_data
```

```python
# 다중회귀분석
import statsmodels.formula.api as sm

result = sm.ols(formula = 'label ~ Area_No + School_No + Exam_No + Area_ID + Exam_ID + Sido_No+ Region_No + School + Age + Gender + Strata + Weight + n_tooth_d + dt_d + ft_d + dft_d + ds_d + fs_d + dfs_d + prev_caries_d + act_caries_d', data = oversampling_data).fit()
result.summary()
```

```python
# p_value 값이 유의미한 값 (0.05보다 작은값) 및 t 값이 큰 값 추출

extraction_data = oversampling_data[['School_No', 'School', 'dt_d', 'dft_d', 'prev_caries_d']]

extraction_data.to_csv("p_value값이 0.05보다 작은 변수 추출.csv")
data = pd.read_csv("p_value값이 0.05보다 작은 변수 추출.csv", index_col = 0)
data
```

```python
# 컬럼명 변경
data.rename(columns={'School_No' : '기관번호', 'School' : '기관구분', 'dt_d' : '우식 유치수', 'dft_d' : '우식경험유치수', 'prev_caries_d' : '유치우식경험자 여부'}, inplace = True)
data
```
