```python
#pip install  scikit-learn // 사이킷런 설치
#pip install pydotplus
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier # 의사 결정 트리 알고리즘
from sklearn import tree
from IPython.display import Image
import pandas as pd
import numpy as np
import pydotplus 
import os
```

```python
ten_data = pd.read_csv('상관계수높은 변수 추출(컬럼명 변경).csv', index_col = 0)
ten_data
```

```python
result_data = pd.read_csv("충치 발생 확률.csv", index_col = 0)
result_data = result_data * 100
result_data
```

```python
# 우식 경험 유치수
# 빈 데이터 프레임 생성
ten_data_2= pd.DataFrame(columns = ['idx','percent(%)'])
i = 0
for idx in range(3112):
    if ten_data['우식경험유치수'][i] == 0:
        percent = 16
        ten_data_2 = ten_data_2.append(pd.DataFrame([[idx, percent]], columns=['idx','percent(%)']), ignore_index=True)
    elif ten_data['우식경험유치수'][i] <= 3:
        percent = 35
        ten_data_2 = ten_data_2.append(pd.DataFrame([[idx, percent]], columns=['idx','percent(%)']), ignore_index=True)
    elif ten_data['우식경험유치수'][i] <= 6:
        percent = 29
        ten_data_2 = ten_data_2.append(pd.DataFrame([[idx, percent]], columns=['idx','percent(%)']), ignore_index=True)
    else :
        percent = 19
        ten_data_2 = ten_data_2.append(pd.DataFrame([[idx, percent]], columns=['idx','percent(%)']), ignore_index=True)
    i = i + 1
ten_data_2.set_index('idx', inplace=True)
ten_data_2
```

```python
ten_data_3 = pd.concat([ten_data,ten_data_2],axis=1)
ten_data_3
```

```python
# 우식 경험 유치수 머신러닝

#속성과 클래스 분리
X = np.array(pd.DataFrame(ten_data_3, columns=['우식경험유치수']))
y = np.array(pd.DataFrame(ten_data_3, columns=['percent(%)']))
y = y.astype('int')
X_train, X_test, y_train, y_test = train_test_split(X,y)


dt_clf = DecisionTreeClassifier()
dt_clf = dt_clf.fit(X_train, y_train)
dt_prediction1 = dt_clf.predict(X_test)

dt_prediction1

# 예측 정확도
print(accuracy_score(y_test, dt_prediction1))
```
